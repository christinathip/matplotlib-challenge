```python
# Dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# File to Load 
clinicaltrial_file = "data/clinicaltrial_data.csv"
mousedrug_file = "data/mouse_drug_data.csv"
```

```python
# Read the Clinical Trial Data and the Mouse and Drug Data
clinicaltrial_df = pd.read_csv(clinicaltrial_file)
mousedrug_df = pd.read_csv(mousedrug_file)

# Combine the data into a single dataset
merged_df = pd.merge(clinicaltrial_df, mousedrug_df, on=('Mouse ID'))

# Display the data table for preview
merged_df.head()
```

```python
# Tumor Response to Treatment
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
mean_vol = pd.DataFrame(merged_df.groupby(['Drug','Timepoint']).mean()['Tumor Volume (mm3)']).unstack(0)
sem_vol = pd.DataFrame(merged_df.groupby(['Drug','Timepoint']).sem()['Tumor Volume (mm3)']).unstack(0)
# Preview DataFrame
mean_vol.head()
```

```python
sem_vol.head()
```

```python
plt.figure(figsize = (20,15))
x_axis = np.arange(0, 50, 5)
plt.xlim = (0, 45)
plt.xticks(np.arange(0, 50, 5)) # location of separators for timepoint
count = 0
for c in mean_vol.columns:
    plt.errorbar(mean_vol.index,
                 mean_vol[c],
                 sem_vol[c],
                 marker =  'o',
                 label = c[1])
count += 1
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)
plt.savefig("data/Response.png")
plt.show()
```

```python
# Metastatic Response to Treatment
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
mean_meta = pd.DataFrame(merged_df.groupby(['Drug','Timepoint']).mean()['Metastatic Sites']).unstack(0)
sem_meta = pd.DataFrame(merged_df.groupby(['Drug','Timepoint']).sem()['Metastatic Sites']).unstack(0)
# Preview DataFrame
mean_meta.head()
```

```python
sem_meta.head()
```

```python
# Generate the Plot (with Error Bars)
# Minor Data Munging to Re-Format the Data Frames
# Preview that Reformatting worked
plt.figure(figsize = (15,10))
plt.xticks(np.arange(0, 50, 5))
plt.xlim = (0, 45)
plt.ylim(0, mean_meta.max().max() + sem_meta.max().max() + .1)

count = 0
xvals = mean_meta.index
# Loop through each Drug
for c in mean_meta:
    plt.errorbar(xvals,
                mean_meta[c],
                sem_meta[c],
                linestyle = '--',
                marker =  'o',
                label = c[1])                                 
count += 1
plt.title("Metastatic Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Metastatics Sites")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)
# Save the Figure
plt.savefig("data/Metastatic.png")
# Show the Figure
plt.show()
```

```python
# Survival Rates
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
mice = pd.DataFrame(merged_df.groupby(['Drug','Timepoint']).count()['Mouse ID']).unstack(0)
# Preview DataFrame
mice.head()
```

```python
plt.figure(figsize = (20,15))
plt.xticks(np.arange(0, 50, 5))
plt.xlim = (0, 45)
xvals = mice.index
count = 0
for c in mice:
    yvals = mice[c]/mice.loc[0,c] * 100
    plt.plot(xvals,
             yvals,
             linestyle = '--',
             marker =  'o',
             label = c[1])
count += 1
plt.title("Survival Rates")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate Percentage")
plt.legend(loc="best", fontsize="small", fancybox=True)
plt.grid('on', which='major', axis='both', linestyle='dotted', linewidth=0.5)
plt.savefig("data/Survival.png")
plt.show()
```
